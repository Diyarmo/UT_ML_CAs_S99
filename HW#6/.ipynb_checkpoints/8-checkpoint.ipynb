```python
import numpy as np
```

```python
def gaussian_signal(x, m, s):
    return (s * (2*np.pi)**.5)**-1 * np.exp(-0.5*(((x - m)/s))**2)
```

```python
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-10, 10, 1000)
```

```python
plt.figure(figsize=(12, 8))
S = []
for m, s in [[0, 1], [1, 2], [-1, .5]]:
    S.append(gaussian_signal(t, m, s))
    plt.plot(S[-1])
S = np.array(S).T
```

همانطور که در تصویر بالا قابل مشاهده است ۳ سیگنال گوسی با میانگین و انحراف معیار‌های مختلف تولید شده اند.

این سه سیگنال سپس به صورت های مختلف با هم ترکیب میشوند.

```python
A = np.array([
    [1, 1, 1],
    [1, .5, .5],
    [0.5, 1, 2]])
```

```python
X = (S @ A.T).T
```

```python
def center(X):
    means = X.mean(axis=1, keepdims=True)
    centered = X - means
    return centered, means

def whiten(X):
    cov = X.dot(X.T) / (X.shape[1] - 1)
    U, S, V = np.linalg.svd(cov)
    D = np.diag(1 / (S)**.5)
    whiteM = np.dot(U, np.dot(D, U.T))
    Xw = np.dot(whiteM, X)
    return Xw, whiteM

def fastICA(X,  thresh=1e-8, iterations=5000):
    m, n = X.shape
    W = np.random.rand(m, m)
    for c in range(m):
            w = W[c, :].copy().reshape(m, 1)
            w = w / np.sqrt((w ** 2).sum())

            lim = 100
            for i in range(iterations):
                if (lim < thresh):
                    break
                wx = np.dot(w.T, X)
                g = np.tanh(wx).T
                g_prime = (1 - np.square(np.tanh(wx)))
                wNew = (X * g.T).mean(axis=1) - g_prime.mean() * w.squeeze()
                wNew = wNew - np.dot(np.dot(wNew, W[:c].T), W[:c])
                wNew = wNew / np.sqrt((wNew ** 2).sum())
                lim = np.abs(np.abs((wNew * w).sum()) - 1)
                w = wNew

            W[c, :] = w.T
    return W
def unmix(Xw, W, means):
    unMixed = Xw.T.dot(W.T)
    unMixed = (unMixed.T - means)
    return unMixed

centered, means = center(X)
Xw, whiteM = whiten(centered)
W = fastICA(Xw)
unMixed = unmix(Xw, W, means)
```

```python
for x in unMixed:
    plt.plot(x)
    plt.show()
```

پس از اجرای الگوریتم مشاهده می شود که اصلا الگوریتم مناسب عمل نکرده است و نتوانسته است سه سیگنال را دوباره تولید کند، چرایی این موضوع به طول کامل در اسلایدها توضیح داده شده است.
